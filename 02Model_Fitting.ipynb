### RFE Recursive Feature Elimination

```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Load the dataset
data = pd.read_csv("expanded.csv")
#data = pd.read_csv("selected_featuresRFEXGB128-2.csv")

# Separate features and labels
features = data.iloc[:,1:-1]  # Exclude the first and last columns
print(features.shape) #shall be 496 columns
labels = data.iloc[:, -1]     # Last column as the label

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Create a random forest classifier
rf = RandomForestClassifier()
estimator = LGBMClassifier(
    #n_estimators=1000,
    #max_depth=6,
    #learning_rate=0.01,

)  # Customize parameters as needed

# Create the RFE object with desired number of features to select
num_features_to_select = 300  # Number of features to select
step_size = 1                 # Step size for RFE
rfe = RFE(estimator=estimator, n_features_to_select=num_features_to_select,step=step_size)

# Perform recursive feature elimination
selected_features = rfe.fit_transform(scaled_features, labels)

# Get the selected feature indices
selected_indices = rfe.get_support(indices=True)

# Get the selected feature names
selected_feature_names = features.columns[selected_indices]

# Create a DataFrame with the selected features and the class label
selected_data = pd.DataFrame(selected_features, columns=selected_feature_names)
selected_data[labels.name] = labels
selected_data.insert(0, 'ID', range(1, len(selected_data) + 1))

# Add ID column as first column
#selected_data.insert(0, 'ID', range(1, 1 + len(selected_data)))

# Save the DataFrame to a new CSV file
selected_data.to_csv("selected300.csv", index=False)
```

### XGB ExtremeGradientBoosting

```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import log_loss, accuracy_score, precision_score, recall_score, f1_score, fbeta_score
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Load the dataset
df = pd.read_csv('selected300.csv')

# Separate features and labels
X = df.iloc[:,1:-1]
y = df.iloc[:, -1]

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Convert the dataset to DMatrix format for XGBoost
train_data = xgb.DMatrix(X_train_smote, label=y_train_smote)
test_data = xgb.DMatrix(X_test, label=y_test)

params = {
    'objective': 'binary:logistic',
    'learning_rate': 0.05,
    'max_depth': 11,
    'min_child_weight': 1,
    'eval_metric': 'logloss',
    'seed': 42
}

num_boost_round = 120000

# Train the model
model = xgb.train(
    params,
    train_data,
    num_boost_round=num_boost_round,
    early_stopping_rounds=50,
    evals=[(test_data, 'validation')]
)

# Predict the labels
y_pred = model.predict(test_data)
y_pred_labels = (y_pred >= 0.5).astype(int)

# Calculate log loss
logloss = log_loss(y_test, y_pred_labels)
print("Log Loss:", logloss)

# Calculate other evaluation metrics (e.g., accuracy, precision, recall, F1 score)
XGB_accuracy = accuracy_score(y_test, y_pred_labels)
precision = precision_score(y_test, y_pred_labels)
recall = recall_score(y_test, y_pred_labels)
f1 = f1_score(y_test, y_pred_labels)

print("Accuracy:", XGB_accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Set the threshold for class assignment
threshold = 0.5

# Convert the test data to DMatrix format for XGBoost
X_test_dmatrix = xgb.DMatrix(X_test)

# Predict the probabilities
y_pred_prob = model.predict(X_test_dmatrix)

# Convert probabilities to class labels
y_pred_labels = (y_pred_prob >= threshold).astype(int)

# Calculate the F2 score
XGB_f2_score = fbeta_score(y_test, y_pred_labels, beta=2, average='weighted')

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_labels)

# Get feature importance
XGB_feature_importance = model.get_score(importance_type='gain')

# Now XGB_feature_importance is a dictionary with feature names as keys.
# Convert it to a list to be able to use argsort().
features = list(XGB_feature_importance.keys())
importances = list(XGB_feature_importance.values())

# Use argsort on importances
sorted_indices = np.argsort(importances)[::-1] # [::-1] to sort in descending order

XGB_sorted_features = [features[i] for i in sorted_indices]
XGB_sorted_importance = [importances[i] for i in sorted_indices]

# Report feature importance
print("Feature Importance:")
for feature, importance in zip(XGB_sorted_features, XGB_sorted_importance):
    print(f"{feature}: {importance}")

# Report the F2 score and accuracy
print("F2 Score:", XGB_f2_score)
print("Accuracy:", XGB_accuracy)
```

## LGBM Hyperparameter tuning (Using BayesianOptimization)

```python
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import log_loss, fbeta_score, accuracy_score
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from bayes_opt import BayesianOptimization

# Load the dataset
df = pd.read_csv('selected300.csv')

# Separate features and labels
X = df.iloc[:,1 :-1]
y = df.iloc[:, -1]

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Set up dataset for LightGBM
train_data = lgb.Dataset(X_train_smote, label=y_train_smote)
valid_data = lgb.Dataset(X_test, label=y_test)

# Define the base parameters
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'binary_logloss',
    'verbose': 0,
    'learning_rate': 0.005,
    #'feature_fraction': 0.8,
    'num_leaves': 80,
    'lambda_l2': 0.02,
}
''''''
# Hyperparameter tuning function
def lgbm_evaluate(bagging_fraction, bagging_freq, feature_fraction):
    params['bagging_fraction'] = bagging_fraction
    params['bagging_freq'] = int(bagging_freq)
    params['feature_fraction'] = feature_fraction

    model = lgb.train(
        params,
        train_data,
        valid_sets=valid_data,
        num_boost_round=120000,
        early_stopping_rounds=100,
        verbose_eval=False
    )

    y_pred = model.predict(X_test)
    y_pred_labels = (y_pred >= 0.5).astype(int)

    LGBM_f2_score = fbeta_score(y_test, y_pred_labels, beta=2, average='weighted')
    LGBM_accuracy = accuracy_score(y_test, y_pred_labels)

    # Save the hyper-parameter set along with their f2 score and accuracy score into a single row csv
    if LGBM_f2_score > 0.7:
        param_values = [bagging_fraction, int(bagging_freq), feature_fraction, LGBM_f2_score, LGBM_accuracy]
        param_names = ['bagging_fraction', 'bagging_freq', 'feature_fraction','f2_score', 'accuracy']
        param_dict = dict(zip(param_names, param_values))
        param_df = pd.DataFrame(param_dict, index=[0])
        param_df.to_csv(f"parameters_{LGBM_accuracy}_LGBM.csv", index=False)

    return LGBM_accuracy

# Apply Bayesian Optimization
lgbm_bo = BayesianOptimization(
    lgbm_evaluate,
    {'bagging_fraction': (0.5,1), 'bagging_freq': (1, 10),'feature_fraction': (0.5,1) }
)

lgbm_bo.maximize(
    init_points=20,
    n_iter=20000
)
```
